Plot Labels

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
#adding additional parameters to the plot 
ax.set_title('Total growth over time') #title
ax.set_ylabel('Total growth') #y axis label
ax.set_xlabel('Years since start') #x axis label
fig.tight_layout() #this adjusts the plot to fit the figure area with the labels - so they don't get cut off
```

```python
#modify the labels using fontdict or font dictionary which is a dictionary of font properties such as font, colour, size, etc.
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}
```

```python
#adding font parameters to the previous line plot

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
# make title left aligned using loc parameter

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1, loc='left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

Customizing Plot Appearance

```python
#Here we remake the basic scatter plot but add colour and marker argument (note colour must use american spelling - color)

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter( x, 
           y, 
           marker='*', # NEW ADDITIONS the maker is changed to a start
           color = "indigo") # NEW ADDITIONS colour is changed to indigo
fig.show()
```

```python
#Here we adjust this previous plot by adding line and adjusting both the line style and width
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot( x, 
           y, 
           marker ='*',
           color = 'indigo',
           linestyle = '--', # NEW ADDITIONS line style is dashed --
           linewidth = 2) # NEW ADDITIONS line width is 2
fig.show()
```

```python
#we can also use specific colour names/hex codes: see: https://matplotlib.org/stable/gallery/color/named_colors.html#sphx-glr-gallery-color-named-colors-py 

fig, ax = plt.subplots(figsize=(5, 3))

# color being the new addition
ax.plot(x, y, marker = '*', color = '#7425b9', linestyle = "--", linewidth = 2) #hex code for purple #7425b9

fig.show()
```

```python
#more customization marker size and colour

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*', 
        markersize = 12, # NEW ADDITIONS here marker size is changed to 12 (larger)
        color = '#7425b9', 
        linestyle = '--', 
        linewidth = 2, 
        markeredgecolor = '#fa9359', # NEW ADDITIONS here the colour for the marker edge is set to #fa9359
        markerfacecolor = '#fa9359' ) # NEW ADDITIONS here the colour for the marker face is set to #fa9359
fig.show()
```

```python
#same as previous exccept the face colour is set to black
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*', 
        markersize = 12, # NEW ADDITIONS 
        color = '#7425b9', 
        linestyle = '--', 
        linewidth = 2, 
        markeredgecolor = '#fa9359', # NEW ADDITIONS 
        markerfacecolor = '#000000' ) # NEW ADDITIONS face colour is set to black
fig.show()

```

```python
#Adding gridlines to the plot

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*', 
        markersize = 12,
        color = '#7425b9', 
        linestyle = '--', 
        linewidth = 2, 
        markeredgecolor = '#fa9359',
        markerfacecolor = '#000000' )
fig.show()

ax.grid(axis = 'y') # NEW ADDITIONS this adds gridlines to the y axis
```

```python
#Here we modify the colour and style of the gridlines 
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*', 
        markersize = 12,
        color = '#7425b9', 
        linestyle = '--', 
        linewidth = 2, 
        markeredgecolor = '#fa9359',
        markerfacecolor = '#000000')
fig.show()

ax.grid(axis = 'y', color = "blue", linewidth = 2, linestyle = '-.')   # NEW ADDITIONS here we specify blue as the line colour, 2 as the line width and, "-." as the line style
```

```python
#Making sample plots - using the below  paramaeters for random seed and integer values

np.random.seed(613) #for reproducibility

x = np.arange(50) 

y1 = np.random.randint(0, 100,50)

y2 = np.random.randint(0, 100,50)
```

```python
#make a simple line plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1,y2) #line plot - has both y1 and y2 on the same plot
```

```python
#from lecture notes: same as above but split into 2 plot arguments - better for formating later

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1)
ax.plot(x,y2)
fig.show()
```

```python
# Adding a legend to the plot - this is done by adding a label within each plot argument and adding the legend fucntion with the location specified at the lower right

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='lower right')
fig.show()
```

```python
#Modifying the legend

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right', frameon = True, #add frame around the legend with frameon = True
          fontsize = 12, #change font size to 12
          ncol = 2, #specify number of columns to 2 for the legend 
          shadow = True ) #add shadow to the legend
fig.show()
```

```python
#Moving the legend box - and moving out of the plot area itself
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc = 'upper left', 
          bbox_to_anchor =(1, 1)) #this specifies the location of the legend box to the upper left outside the pot area and the anchor point is set to (1,1) which is the top right corner
fig.show()
```

```python
#Moving the legend box - and moving out of the plot area itself - another example
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc = 'upper left', 
          bbox_to_anchor =(1, 0.5)) #changed to middle of the y axis
fig.show()
```

```python
#Moving the legend box - and moving out of the plot area itself - another example
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(#loc = 'upper left', #removed this to see if it still works without it
          bbox_to_anchor =(0.5, 0.5)) #changed to middle of the x and y axis
fig.show()
```

Text and Annotations

```python
#Modifying with text annotations - now using a scatter plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
```

```python
#Using text function we can specify text coordinates and text itself
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.text(10, 95, "This value is important!")
fig.show()
```

```python
#Like with every other modification we can modify the text
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.text(10, 95, "This value is important!", 
        ha='center', #alignment of text to center
        color = 'red', #modify font colour to red
        size = 20) #modify font size to 20

fig.show()
```

```python
#Positioning our text annotations - using transforms

#Start with a set of empty axes
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])

#We now can position our annotation with ref to the data (as we did before)
#Or we can do this according to axes or the figure itself (referenceing the axis or figure dimensions)

ax.text(1, 5, ". Data:(1, 5)", transform=ax.transData)
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform=ax.transAxes)
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform=fig.transFigure)

```

```python
#Annotating with arrows - using the annotate function

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94), #the arrow points at (10, 95) and the text is at (20, 94)
            arrowprops=dict(facecolor='black', shrink=0.05)) #we can modify this to make changes to arrow style and colour)
fig.show()
```

```python
#Annotating with arrows - using the annotate function

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94), #the arrow points at (10, 95) and the text is at (20, 94)
            arrowprops=dict(arrowstyle = "wedge", color='hotpink')) #we can modify this to make changes to arrow style and colour)
fig.show()
```

Axis Markers/Labels

```python
# We can remove tick marks and labels

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right') 
ax.yaxis.set_major_locator(plt.NullLocator())  #remove y axis ticks and labels
ax.xaxis.set_major_formatter(plt.NullFormatter()) #remove x axis ticks and labels
```

```python
#Limiting the tick marks
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3)) #this limits the x axis to 3 tick marks
```

```python
#Limiting the tick mark intervals
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) #this sets the x axis to intervals of 5
```

```python
#Rotating Axis Labels - via a degree of rotation and horizontal alignment
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

plt.xticks(rotation=45, ha='right') #this rotates the x axis labels by 45 degrees with right allignment
```
